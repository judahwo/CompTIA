## OSI Model 1.1

**(Open Systems Interconnection)**<br>
- Layer 7: Application
- Layer 6: Presentation
- Layer 5: Session
- Layer 4: Transport
- Layer 3: Network
- Layer 2: Data Link
- Layer 1: Physical




### Layer 1: Physical
- The Physics of the Network
    - This layer isn't about protocos, it's about the physical devices that connect to the network.

- Layer 1 Devices
    - Cables
    - Hubs
    - Fibers
    - The signal itself


### Layer 2: Data Link

- The basic network "language"
    - The foundation of communiacation at the data link layer
    - The data link layer is responsible for the physical addressing of devices

- Data Link Control (DLC) protocols
    - Ethernet
    - MAC addresses
    - Wireless
    - Bluetooth
    - Wi-Fi
    - Switches



### Layer 3: Network

- The network layer is responsible for routing
    - IP addresses
    - Routers
    - Packet

### Layer 4: Transport

- The transport layer is responsible for end-to-end communication
    - TCP
    - UDP
    - Ports
    - Sockets

### Layer 5: Session

- The session layer is responsible for establishing, maintaining, and terminating connections
    - Control protocols
    - Tunneling protocols
    

### Layer 6: Presentation

- The presentation layer is responsible for data translation and encryption
    - Data formats
    - Encryption
    - Compression
    - SSL/TLS

### Layer 7: Application

- The layer where you interact with the application

**Summary**

- Data Link: Ethernet
- Network: IP encapsulation
- Transport: TCP encapsulation
- Session: Link the presentation to the transport
- Presentation: SSL encryption

## Network Devices 1.2

**Link:** [Network Devices](https://www.youtube.com/watch?v=iqjj4ZSPV08)

### Router
- Routes traffic between IP subnets
    - Layer 3 device
    - Connects different networks
    - Uses IP addresses


### Switch

- Connects devices on the same network
    - Layer 2 device
        - Forwards traffic based on MAC addresses
    - Uses MAC addresses
    - Broadcasts traffic to all devices
    - Bridging done in hardware
        - ASICs (Application Specific Integrated Circuits)
    - Many ports and features
        - The core of an enterprise network
        - May provide PoE (Power over Ethernet)
    - Multilayer switches
        - Layer 3 routing


### Firewall

- Traditional
    - Blocks traffic based on rules
    - Stateful inspection
    - NAT (Network Address Translation)
    - VPN (Virtual Private Network)
- Next-Generation
    - Application awareness
    - Deep packet inspection
    - User identification
    - Threat intelligence
- Encrypt traffic
    - VPN between sites
- Most firewalls can be 3 layer devices
    - often sites on ingress/egress of network
    - dynamic routing
    


### IDS/IPS

- Intrusion Detection System (IDS)
    - Detects attacks
    - Logs events
    - Alerts
- Intrusion Prevention System (IPS)
    - Detects attacks
    - Logs events
    - Alerts
    - Blocks traffic
- Intrusions
    - Exploits against operating systems, applications, or services
    - Buffer overflows, cross-site scripting, other vulnerabilities
- Detection vs. Prevention
    - Dectection: Log and alert
    - Prevention: Log, alert, and block

### Load Balancer

<br>
- Distributes traffic across multiple servers
    - Redundancy
    - Scalability
    - Performance
- Layer 4
    - TCP/UDP
    - Source IP, destination IP, source port, destination port
- Large scale implementations
    - web server farms
    - database clusters
- Fault tolerance
    - server outages have no effect
    - very fast convergecne 
- Configurable Load
    - manage across servers
- TCP offloading
    - protocol overhead
- SSL offloading
    - encryption/decryption
- Caching
    - faster response times

### Proxy Server

<br>

- Between users and the external network
    - Web proxy
    - Forward proxy
    - Reverse proxy
- Receives the user request and sends the request on their behalf (the proxy)
- Caches content
    - Faster access
    - Less bandwidth
- Filters content
    - URL filtering
    - Malware scanning
    - Content inspection
- Explicit vs. Transparent
    - Explicit: User configures the proxy
    - Transparent: User doesn't know about the proxy

### NAS vs. SAN

- Network Attached Storage (NAS)
    - File-level storage
    - Network-based
    - File sharing
    - SMB, NFS, FTP
- Storage Area Network (SAN)
    - Block-level storage
    - Network-based
    - Direct access to storage
    - iSCSI, Fibre Channel

### Access Point (AP)

- Not a router
    - an access point is a bridge between wireless and wired networks
- Wireless access point
    - extends the wired network to wireless devices
- OSI Layer 2 device
    - MAC addresses
    - Wireless encryption
    - SSID
    - WPA2, WPA3

### Wireless LAN Controller (WLC)

- Centralized management
    - Configuration
    - Monitoring
    - Security
- Deploy new access points
    - Automatically configure
    - Automatically update
- Performance and Security Monitoring
    - Rogue AP detection
    - Wireless intrusion prevention
    - Spectrum analysis
- Configure and deploy changes to all sites
- Report on access point use

### Content Delivery Network (CDN)

    - Bandwidth

- Geographically distributed servers

### Virtual Private Network (VPN)

- Secure communication over an untrusted network
    - Encryption
    - Authentication
    - Integrity
- Concentrator / headend
    - Terminates VPN connections
    - Encryption / decryption access device
    - Often integrated with firewalls
    - Many VPN connections
- Many deployment options
    - Site-to-site
    - Remote access
    - Client-to-site
    - Client-to-client
- Often used with client software
    - VPN client
    - VPN concentrator
    - VPN gateway
    - Sometimes built into the operating system

### Quality of Service (QoS)

- Traffic Shaping
    - Control the flow of traffic
    - Prioritize traffic
- Packet Shaping
    - Control the flow of packets
    - Prioritize packets
- Control Bandwidth
    - Limit bandwidth
    - Guarantee bandwidth
- Prioritize Traffic
- Manage the QoS
    - routers
    - switches
    - firewalls
    - other network devices

### Time to live (TTL)

- How long should data be available?
    - Not all systems or protocals are self regulating
    - Sometimes we need to set a limit
- Create a timer
    - Set a limit
    - When the timer expires, the data is no longer valid
- Many different uses
    - drop packets
    - expire data
    - clear cache

### Routing Loops

- Route A thinks Route B is the best path
    - Route B thinks Route A is the best path
    - The traffic goes back and forth
- Easy to misconfigure
    - Especiatlly with static routing
- This can't go on forever
    - TTL (Time to Live) will eventually expire
    - The packet will be dropped

### Internet Protocol (IP)

- Loops could cause a packet to live forever
    - drop the packet after a certain number of hops
- Each pass through a router is a hop
    - Default TTL for macOS is 64
    - Default TTL for Windows is 128
- The router decrements the TTL
    - When the TTL reaches 0, the packet is dropped

### Domain Name System (DNS)

- DNS lookups
    - resolve names to IP addresses
    - www.professormesser.com = 172.67.41.114
- A device queries a DNS server
    - The DNS server responds with the IP address

    - Broadcast Address
        - 192.168.1.255

    - Allows many devices to share a single public IP address
    - Many devices can use the same private IP address
    - NAT is a function of the router
