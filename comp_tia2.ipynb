### Network Components

**Course** [Network Components](https://www.udemy.com/course/comptia-network-009/learn/lecture/42185520#content)

**Clients**<br>
- Devices that users access the network with
- Examples: desktops, laptops, tablets, smartphones, printers, etc.

**Servers**<br>
- Provide resources to the network
- Examples: file servers, print servers, web servers, etc.

**Hub**<br>
- Older netword devices
- Connects multiple devices like clients and servers over a local area network (LAN)
- Hubs evovled into brides, and then switches over time

**Switch**<br>
- "Smarter Hubs"
- Provide more security and efficient bandwidth utilization

**Wireless Access Point (WAPs/APs)**<br>
- Allow wireless devices to connect to the wired network

**Router**<br>
- Connects multiple networks together
- Routes traffic between networks

**Firewall**<br>
- Protects the network from unauthorized access
- Filters traffic based on rules

**Load Balancer**<br>
- Distributes traffic across multiple servers
- Ensures no single server is overwhelmed

**Proxy Server**<br>
- Acts as an intermediary between clients and servers
- Caches web pages to improve performance
- Web filtering
- Shared network connections
- Data caching
- Enhanced security and privacy

**Intrustion Detection System (IDS)**<br>
- Monitors network traffic for suspicious activity
- Alerts administrators to potential security threats

**Intrustion Prevention System (IPS)**<br>
- Monitors network traffic for suspicious activity
- Can take action to prevent security threats

**Controller**<br>
- Manages multiple access points in a wireless network
- Provides centralized management and configuration

**Network Attached Storage (NAS)**<br>
- Storage device that connects to the network
- Provides file storage to clients and servers

**Storage Area Network (SAN)**<br>
- High-speed network that connects storage devices to servers
- Provides block-level storage to servers

**Media**<br>
- Physical layer of the network
- Examples: copper, fiber, wireless

**Wide Area Network (WAN) Technologies**<br>
- Connects multiple LANs together
- Examples: T1, T3, DSL, cable, fiber, etc.





### Network Resources

**Course** [Network Resources](https://www.udemy.com/course/comptia-network-009/learn/lecture/42185522#content)

**Client Server Model**<br>
- Clients request resources from servers
- Servers provide resources to clients
- Administration and management
    - Centralized administration
    - Easier management
    - Scalability
- Some downsides
    - Costs more
    - Requires dedicated OS
    - Requires specialized skillset

**Peer-to-Peer Model**<br>
- Clients share resources with each other
- No centralized server
- Benefits
    - Lower cost
    - No dedicated OS
    - No dedicated resources
- Some downsides
    - Security concerns
    - Inefficient for large networks
    - Decentralized management
    - Scalability issues

A **peer-to-peer network** model does not differentiate between the clients and the servers, and every node can become a client and a server when requesting and responding to service requests. In a **client-server network model**, there is a clear distinction between client devices that request services and server devices that provide services. A **Point-to-point network** is a direct connection between two devices. One device acts as the sender and the other as the receiver. A **Hub-and-spoke network** model has a central hub that connects other devices (spokes). The hub provides services to the spokes.





### Network Geography

**Course** [Network Geography](https://www.udemy.com/course/comptia-network-009/learn/lecture/42185524#content)

**Personal Area Network (PAN)**<br>
- Smallest type of network
- Usually covers a distance of 10 meters or less
- Examples: Bluetooth, USB

**Local Area Network (LAN)**<br>
- Covers a small geographic area
- Usually confined to a single building or campus
- Usually covers a distance of 100 meters or less
- Examples: Ethernet, Wi-Fi
    - Wi-Fi: IEEE 802.11
    - Ethernet: IEEE 802.3

**Campus Area Network (CAN)**<br>
- Covers a larger geographic area
- Usually covers a distance up to several kilometers
- Examples: college campus, business park, military base

**Metropolitan Area Network (MAN)**<br>
- Covers a city or metropolitan area
- Usually covers a distance up to 50 kilometers
- Examples: mutliple colleges, department of motor vehicles, etc.

**Wide Area Network (WAN)**<br>
- Covers a large geographic area
- Across a state, country, or even the world
- Examples: Internet, MPLS, Frame Relay, etc.

A **campus area network** (CAN) is a computer network that spans a limited geographic area. CANs interconnect multiple **local area networks** (LAN) within an educational or corporate campus. A WAN (**Wide Area Network**) is a network that spans a large geographical area, such as multiple cities or countries. A PAN (**Personal Area Network**) is a network that connects devices within an individual's personal space, typically within a range of a few meters. A MAN (**Metropolitan Area Network**) is a network that spans a larger area than a LAN (**Local Area Network**), often covering a city or a metropolitan area.


### Wired Network Topologies

**Course** [Wired Network Topologies](https://www.udemy.com/course/comptia-network-009/learn/lecture/42185530#content)

**What is a Network Topology?**<br>
- Referes to the physical layout of the network
- Determines how devices are connected
- Determines how data flows through the network
- Examples: links,nodes, client, server, etc.
- Makes up a computer network

**Phyiical Topologies**<br>
- Used to show how devices are physically connected
- Examples: floor plan, building layout, etc.

**Logical Topologies**<br>
- Used to show how data flows through the network
- Examples: logical diagrams, network maps, etc.

**Point-to-Point Topology**<br
- Connects two devices directly
- Examples: serial cable, USB cable, etc.
- Pros: simple, inexpensive, easy to install for small scale
- Cons: limited scalability, single point of failure

**Ring Topology**<br>
- Connects devices in a circular fashion
- Data flows in one direction
- Examples: Token Ring, FDDI, etc.
- Pros: simple, easy to install, inexpensive
- Cons: single point of failure, limited scalability, performance issues

**Bus Topology**<br>
- Connects network devices to a single central cable
- Called a bus or backbone
- Pros: easy to install, inexpensive, requires less cable
- Cons: single point of failure, limited scalability, performance issues

**Star Topology**<br>
- Most common used today
- Connects devices to a central router, hub, or switch
- Connects back to a central point
- If centrail point fails, the network fails
- Pros: good for centralized management
- Cons: single point of failure, relies on central device

**Hub and Spoke Topology**<br>
- Connects multiple star topologies together
- Central hub connects to multiple spokes
- Pros: good for centralized management
- Cons: single point of failure, relies on central device

**Mesh Topology**<br>
- Features a point-to-point connection between every device on the network
- Fully meshed: every device connects to every other device
    - Pros: fault tolerance, redundancy, scalability
    - Cons: expensive, complex, difficult to manage
- Partially meshed: some devices connect to every other device
    - Pros: less expensive, less complex, easier to manage

A **mesh topology** connects every endpoint to every other endpoint, creating a fully redundant network. While a **hybrid topology** combines different topologies, it does not inherently provide full redundancy. The redundancy level depends on the specific combination of topologies used. In a **bus topology**, all devices are connected to a single shared communication line. If a device on it fails, the entire network can be affected. It lacks redundancy. In a **star topology**, devices are connected to a central hub or switch. If the hub or switch fails, the network can be disrupted.

A **hub and spoke topology** is a network topology where a central device (the hub) is connected to multiple other devices (the spokes). A **Bus topology** has a single communication line where all devices are connected. It does not have distinct hubs and spokes. A **star topology** has a central hub that connects all devices in a point-to-point manner. While the Dion Worldwide network has hubs, not a single central hub connecting all devices. A **Ring topology** connects devices in a circular manner, where each device is connected to two neighboring devices. It doesnâ€™t have distinct hubs and spokes.

### Datacenter Architecture

**Course** [Datacenter Architecture](https://www.udemy.com/course/comptia-network-009/learn/lecture/42185534#content)

**Datacenter**<br>
- A facility that houses servers, storage, and networking equipment
- Provides centralized resources to the network
- Examples: Google, Amazon, Microsoft, etc.

**3-Tier Hierarchical Model**<br>
- Core Layer: high-speed backbone
    - biggest, fastest, most expensive
- Distribution Layer: connects core to access layer
    - connects core to access layer
- Access Layer: connects clients to the network
    - connects clients to the network

**Collapsed Core Model**<br>
- Core and distribution layers are combined into one layer
- Simplifies the network design
- Reduces the number of devices
- May not be as scalable as the 3-tier model, not suitable for large or complex networks

**Spine-Leaf Model**<br>
- Alternative type of network architecture
- Focuses on communication within the datacenter itself
- faster speeds and lower latency than 3-tier model
- Spine: connects switches in a full mesh topology
- Leaf: consists of all the access switches

**Traffic Flow**<br>
- North-South: refers to traffic flow that enters or leaves the datacenter from a system phycially outside the datacenter
    - North: traffic exits the datacenter
    - South: traffic enters the datacenter
    - Traffice enters/exits through firewall or router
- East-West: refers to traffic flow within the data center
    




## OSI Model

**Course** [OSI Model](https://www.udemy.com/course/comptia-network-009/learn/lecture/42185536#content)

**OSI Model**<br>
- Open Systems Interconnection Model
- Developed by the International Organization for Standardization (ISO)

**Domain 1: Networking Concepts**<br>
- Objective 1.1: Explain concepts related to the Open Systems Interconnection (OSI) model

**Reference Model**<br>
- Used to categorize the functions of a network in the OSI Model during troubleshooting efforts
- Networks today operate under TCP/IP model

**OSI Model Layers**<br>
- Please Do Not Throw Sausage Pizza Away
- Physical, Data Link, Network, Transport, Session, Presentation, Application

1. **Physical Layer**<br>
    - Layer 1
    - Transmits raw data bits over a physical medium
    - Examples: cables, connectors, switches, etc.

2. **Data Link Layer**<br>
    - Layer 2
    - Provides error detection and correction
    - Examples: switches, bridges, NICs, etc.

3. **Network Layer**<br>
    - Layer 3
    - Provides logical addressing and routing
    - Examples: routers, IP addresses, etc.

4. **Transport Layer**<br>
    - Layer 4
    - Provides end-to-end communication
    - Examples: TCP, UDP, etc.

5. **Session Layer**<br>
    - Layer 5
    - Establishes, maintains, and terminates connections
    - Examples: NetBIOS, PPTP, etc.

6. **Presentation Layer**<br>
    - Layer 6
    - Translates data into a format that can be understood by the application
    - Examples: encryption, decryption, etc.

7. **Application Layer**<br>
    - Layer 7
    - Provides network services to applications
    - Examples: HTTP, FTP, etc.
    


### Layer 1: Physical Layer

**Course** [Physical Layer](https://www.udemy.com/course/comptia-network-009/learn/lecture/42185546#content)

**Physical Layer**<br>
- where bits are transmitted over a physical medium
- includes physcial and elecletrical network components
- examples: ethernet, fiber, copper, etc.
- Layer 1 devices view networks from a physical topology perspective

**Transition Modulation**<br>
- If it changes during the clock cycle, then 1 is represented, otherwise 0 is represented

**Synchronizing Communication**<br>
- Asynchronous: no clock signal
    - uses start and stop bits
    - indicates when transmissions occur from the sender to the receiver
- Synchronous: clock signal
    - uses a clock signal to synchronize transmissions
    - data is sent in a continuous stream

**Bandwidth**<br>
- Broadband: high-speed data transmission
    - divides bandwidth into multiple channels
- Baseband: single channel data transmission
    - all frequency of the cable all of the time
    - uses a reference clock

**Time-Division Multiplexing (TDM)**<br>
- Each session takes a turn transmitting data
- Shares the medium between all usesrs via time slots

**Statistical Time-Division Multiplexing (STDM)**<br>
- Dynamically allocates the time slots on an as-needed basis
- More efficient than TDM

**Frequency-Division Multiplexing (FDM)**<br>
- Divides the medium into channels based on frequesncy
- Each session is transmitted on a different channel

**Multiplexing**<br>
- involves optimizing a limited amout of resources for more efficient utilization

**Examples of Layer 1 Devices**<br>
- fiber optic cable
- copper cable
- ethernet cable
- coaxial cable
- bluetooth
- wifi
- near field communication (NFC)
- Hubs
- Access Points
- Media converters

Layer 1 devices have no logic or intelligence, they simply repeat what they are told. They are essentially repeaters passing along whatever is received. 




### Layer 2: Data Link Layer

**Course** [Data Link Layer](https://www.udemy.com/course/comptia-network-009/learn/lecture/42185548#content)

**Data Link Layer**<br>
- Packages data into frames
- Transmits those frames on the network

**MAC Address**<br>
- Media Access Control Address
- Unique identifier for network devices
- Physical addressing system of a device
- Operates on a logical topology
- every manufacturer assigns a unique MAC address to each device
- 48-bit address
- 24-bit Vendor ID
- 24-bit Unique ID

**Logical Link Control (LLC)**<br>
- Provides connection services and allows acknowledgement of receipt messages
- Most basic form of flow control
- Gives basic error control functions

**Communication Across Layer 2**<br>
- Isochronous: devices use common reference clock and create time slots for transmission
- Syncronous: devices agree on clocking method to indicate beginning and end of frames and can use control character 
- Asynchronous: devices reference internal clocks and use start and stop bits for synchronization

**Layer 2 Devices**<br> 
- Network Interface Cards (NICs)
- Switches
- Bridges
- MAC Addresses


### Layer 3: Network Layer

**Course** [Network Layer](https://www.udemy.com/course/comptia-network-009/learn/lecture/42185550#content)

**Network Layer**<br>
- Provides logical addressing and routing
- IPV4 and IPV6

**Logical Addressing**<br>
- Logical addressing is used to identify devices on a network
- Examples: IP addresses, MAC addresses, etc.

**Routing**<br>
- Packet Switching
    - data is divided into packets and then forwarded
    - most common form of switching
- Circuit Switching
    - data is sent over a dedicated circuit
    - conastant communication path
- Message Switching
    - data is divided into meassages and then forwarded

**Route Discovery and Selection**<br>
- Manually configued as a static route or dynamically through a routing protocol
- Routing Protocols: RIP, OSPF, EIGRP, BGP, etc.
- Routing prococals help us decide how data is going to flow, and how the routers are going to communicate that information

**Connection Services**<br>
- Augment Layer 2 connection services to improve reliability and efficiency
- Flow control
- Packet reordering
    - ensures all data reaches receiver correctly at the end desitination

**Internet Control Message Protocol (ICMP)**<br>
- ping most commonly used
- sends out packets to check connectivity
- traceroute: checks the path of the packet

**Layer 3 Devices**<br>
- Routers
- Multilayer Switches
    - regular switch and router combined
    - switch is always a layer 2 device unless it is multilayer
- IP Addresses


### Layer 4: Transport Layer
